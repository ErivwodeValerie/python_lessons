Task: Write a python function to find the factorial of any given number

```python
def factorial_param(x):
  while True:
    (x-1) > 0
    return x*(x-1)
    break
print(factorial_param(5))
```

```python
#Solution
def factorial(n):
  if n < 0:
    return "Factorial is not defined for negative numbers"
  result= 1

  for i in range(1, n + 1):
    result*= i
  return result
```

```python
factorial(10)
```
